**Column oriented : Dictionary of lists**

Data is a dict with column names as keys and the respective column values as list of values 

```
cols_dict = {
    'city_name': ['San Francisco', 'San Jose', 'Sacramento'], 
    'population': [852469, 1015785, 485199]
}

pd.DataFrame(cols_dict)
```

**Row oriented : List of dictionary items**

Data is a list of dict items with column names as key and the respective cell value as value. 

```
rows_dict = [{'city_name': 'San Francisco', 'population':852469}, 
             {'city_name': 'San Jose', 'population': 1015785},
             {'city_name': 'Sacramento', 'population': 485199}] 

pd.DataFrame(rows_dict)
```

**Row oriented : List of tuple values**

Data is a list with row values as a tuple and the column names specified as list. 

```
rows_dict = [('San Francisco', 852469), 
             ('San Jose', 1015785),
             ('Sacramento', 485199)] 
labels = ['city_name', 'population']

pd.DataFrame(rows_dict, columns=labels)
```

With different indices specified in `index`

```
pd.DataFrame(cols_dict, index=[104,108,111])
```

**Column oriented : Dictionary of Series**

Data is a dict with column names as keys and the respective column values as `Series` of values. 

```
city_names = pd.Series(['San Francisco', 'San Jose', 'Sacramento'])
populations = pd.Series([852469, 1015785, 485199])

pd.DataFrame({ 'city_name': city_names, 'population': populations })
```

**Saving `DataFrame` to local CSV file**

```
cities_df = pd.DataFrame({ 'city_name': city_names, 'population': populations })
cities_df.to_csv('cities.csv') #index=None
```

DataFrame provides indexing label loc for selecting columns and rows by names to selects the specified columns and rows from the given DataFrame.

`dataFrame.loc[<ROWS RANGE> , <COLUMNS RANGE>]`

ROWS OR COLUMN RANGE can be also be ‘:’  and if given in rows or column Range parameter then the all entries will be included for corresponding row or column.

```
df.loc[:,'Country',]
```

```
df.loc[2:5,['Country','Quantity']]
```

```
df.loc[2:5, 'Sales Person':'Quantity']
```

```
df.loc[11::2]
```

```
df.loc[11,:]
```

```
df.loc[[11,17],:]
```

`DataFrame` provides indexing label `iloc` for accessing the column and rows by index positions to select the columns and rows from DataFrame by index position specified in range. 

`dataFrame.iloc[<ROWS INDEX RANGE> , <COLUMNS INDEX RANGE>]`

If ‘:’ is given in rows or column Index Range then all entries will be included for corresponding row or column.

```
df.iloc[0]
```

```
df.iloc[0,0]
```

```
df.iloc[::2,::2]
```

```
df.iloc[[11,13,17],[2,5]]
```
