Using subplot2grid

````python
import matplotlib.pyplot as plt
import numpy as np

def random_plots():
  xs = []
  ys = []
  
  for i in range(20):
    x = i
    y = np.random.randint(10)
    
    xs.append(x)
    ys.append(y)
  
  return xs, ys

fig = plt.figure()
ax1 = plt.subplot2grid((5, 2), (0, 0), rowspan=1, colspan=2)
ax2 = plt.subplot2grid((5, 2), (1, 0), rowspan=3, colspan=2)
ax3 = plt.subplot2grid((5, 2), (4, 0), rowspan=1, colspan=1)
ax4 = plt.subplot2grid((5, 2), (4, 1), rowspan=1, colspan=1)

x, y = random_plots()
ax1.plot(x, y)

x, y = random_plots()
ax2.plot(x, y)

x, y = random_plots()
ax3.plot(x, y)

x, y = random_plots()
ax4.plot(x, y)

plt.tight_layout()
plt.show()
````

#### 1. Generate a numpy array containing the first 50 natural numbers. Compute another array that is equivalent to the element-wise square root of the first array. 


> #### You should do this in two different ways: <br/>  (i) Using the math.sqrt() function inside a list comprehension computing explicitly for each element through a loop <br/> (ii) Using the numpy ufunc np.sqrt().   


#### Use the python timeit function to compute and print the time taken for performing (i) and (ii). Note down the speedup/slowdown achieved in the case (ii).

````python
import timeit

time_1 = timeit.timeit(setup='import math; import numpy as np; array = np.arange(1,51)',stmt='np.array([math.sqrt(x) for x in array])', number=10000) 
time_2 = timeit.timeit(setup='import numpy as np; array = np.arange(1,51)',stmt='np.sqrt(array)', number=10000) 

print('time_1 = ',time_1,' s')
print('time_2 = ',time_2,' s')

print("Difference  = {:.3f} s".format(time_1-time_2))
print("Speedup    = {:.3f} x".format(time_1/time_2))
print("numpy function %time saved = {:.3f}%".format((time_1-time_2)*100/time_1))
````

Read more about [timeit](https://docs.python.org/3/library/timeit.html) if iterested.

If it is unclear, what happened above, please refer to this code with an implementation of almost the same thing done by timeit

````python
import time
import math
import numpy as np

array = np.arange(1,51)

def timer(function, num_executions = 10000):
  start = time.time()
  for _ in np.arange(num_executions):
    function()
  end = time.time()
  return end - start

time_1 = timer(lambda : np.array([math.sqrt(x) for x in array])) 
time_2 = timer(lambda : np.sqrt(array)) 

print('time_1 = ',time_1,' s')
print('time_2 = ',time_2,' s')

print("Difference  = {:.3f} s".format(time_1-time_2))
print("Speedup    = {:.3f} x".format(time_1/time_2))
print("numpy function %time saved = {:.3f}%".format((time_1-time_2)*100/time_1))
````

#### 5. The parametric function defining a limaon is:

> $$r = r_0 + cos \;$$
> $$x = r \; cos \;$$
> $$y = r \; sin \;$$


#### Plot the shape of the limaon for $r_0 = 0.5$, $r_0 = 1.0$, and $r_0 = 1.5$. 
#### Ensure that you plot enough points that the curve is closed and appears smooth.Use a legend to identify which curve corresponds to which value of $r_0$.         

Note: $x$ and $y$ represent the values of the respective coordinates on the XY-plane. $r$ and $\theta$ are values (parameters) that determine these values $x$ and $y$.  

````python
import numpy as np
from matplotlib import pyplot as plt

def compute_limacon(r0,theta):
  r = r0 + 2 * np.cos(theta)
  x = r * np.cos(theta)
  y = r * np.sin(theta)
  return x,y

theta = np.linspace(0, 2*np.pi, 10000)

r0_values = np.array([0.5,1,1.5,2])

for r0 in r0_values:
  x,y = compute_limacon(r0,theta)
  plt.plot(x,y, label='r0 = {:.1f}'.format(r0))
  
plt.legend()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
x = np.arange(0,11,0.1)
gx = x*np.exp(-x / 3)
fx = np.exp(-x/10)*np.sin(np.pi*x)
plt.plot(fx, label="f(x)")
plt.plot(gx, label="g(x)")
plt.xlabel("x")
plt.ylabel("f(x) / g(x)")
plt.xticks(np.arange(11)*10, [str(i) for i in range(11)])
plt.grid()
plt.legend();
````
