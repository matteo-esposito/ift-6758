**Column oriented : Dictionary of lists**

Data is a dict with column names as keys and the respective column values as list of values 

```python
cols_dict = {
    'city_name': ['San Francisco', 'San Jose', 'Sacramento'], 
    'population': [852469, 1015785, 485199]
}

pd.DataFrame(cols_dict)
```

**Row oriented : List of dictionary items**

Data is a list of dict items with column names as key and the respective cell value as value. 

```python
rows_dict = [{'city_name': 'San Francisco', 'population':852469}, 
             {'city_name': 'San Jose', 'population': 1015785},
             {'city_name': 'Sacramento', 'population': 485199}] 

pd.DataFrame(rows_dict)
```

**Row oriented : List of tuple values**

Data is a list with row values as a tuple and the column names specified as list. 

```python
rows_dict = [('San Francisco', 852469), 
             ('San Jose', 1015785),
             ('Sacramento', 485199)] 
labels = ['city_name', 'population']

#pd.DataFrame(rows_dict, columns=labels)

pd.DataFrame(rows_dict)
```

With different indices specified in `index`

```python
pd.DataFrame(cols_dict, index=[104,108,111])
```

**Column oriented : Dictionary of Series**

Data is a dict with column names as keys and the respective column values as `Series` of values. 

```python
city_names = pd.Series(['San Francisco', 'San Jose', 'Sacramento'])
populations = pd.Series([852469, 1015785, 485199])

pd.DataFrame({ 'city_name': city_names, 'population': populations })
```

**Saving `DataFrame` to local CSV file**

```python
cities_df = pd.DataFrame({ 'city_name': city_names, 'population': populations })
cities_df.to_csv('cities.csv', index=None)
```

#### 5. Cleaning a DataFrame

```python
import numpy as np
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm', 
                               'Budapest_PaRis', 'Brussels_londOn'],
              'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
              'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
                   'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )', 
                               '12. Air France', '"Swiss Air"']})
```

Some values in the the **FlightNumber** column are missing (they are `NaN`). These numbers are meant to increase by 10 with each row so 10055 and 10075 need to be put in place. Modify `df` to fill in these missing numbers and make the column an integer column (instead of a float column).

```python
df['FlightNumber'] = df['FlightNumber'].interpolate().astype(int)
```

The **From\_To** column would be better as two separate columns! Split each string on the underscore delimiter `_` to give a new temporary DataFrame called 'temp' with the correct values. Assign the correct column names 'From' and 'To' to this temporary DataFrame. 

```python
temp = df.From_To.str.split('_', expand=True)
temp.columns = ['From', 'To']
temp
```

Notice how the capitalisation of the city names is all mixed up in this temporary DataFrame 'temp'. Standardise the strings so that only the first letter is uppercase (e.g. "londON" should become "London".)

```python
temp['From'] = temp['From'].str.capitalize()
temp['To'] = temp['To'].str.capitalize()
temp
```

Delete the From_To column from **41.** Delete the **From_To** column from `df` and attach the temporary DataFrame 'temp' from the previous questions.`df` and attach the temporary DataFrame from the previous questions.
